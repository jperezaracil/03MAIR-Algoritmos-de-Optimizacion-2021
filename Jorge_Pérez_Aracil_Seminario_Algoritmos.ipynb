La estructura de datos que, a priori, mejor se adapta al problema es de tipo árbol, pues permite entender el problema de forma sencilla y ordenada. 

Cuando el problema se analiza en profundidad, podemos observar que una estructura tipo array puede adaptarse mejor al problema. En concreto, utilizando la notación "Polish" invertida [1], en la que los símbolos de las operaciones se expresan como sufijos de los operandos, es decir:

$ a + b → a \ b \ +$ 

Esta notación permite, desde el punto de vista del autor de este trabajo, ordenar el conjunto de operaciones posibles. Para ello, sólo es necesario crear una función, que hemos llamado $\textit{calc_RPN}$, que permita interpretar el tipo de operación que se está realizando. 

````
def calc_RPN(lst):
    #Se crea un diccionario que recoge cada una de las posibles operaciones y ejecuta
    # la misma mediante funciones lambda. 
    dc = {
      '+': lambda x, y: x + y, 
      '-': lambda x, y: x - y,
      '*': lambda x, y: x * y, 
      '/': lambda x, y: x / y,
    }
    stack = [] # se crea una lista vacía que permita recoger los operandos 
    resultado = 0
    for i in lst: # se van leyendo cada uno de los elementos de la lista, que contiene tanto operandos como operadores
        if i in dc: # se comprueba si el elemento de la lista es un símbolo
            b, a = stack.pop(), stack.pop() # se sacan esos elementos de la lista y se almacenan en variables
            res = dc[i](a, b) #se calcula el resultado de la operación utilizando las funciones del diccionario
            stack.append(resultado) # se almacena el resultado como un elmento de la lista más
        else: # en caso de no ser un símbolo, se almacena el valor como en la lista 
            stack.append(i) 
    return resultado
````

# Diseña un algoritmo para resolver el problema por fuerza bruta

El algoritmo por fuerza bruta implica analizar todas las posibles permutaciones de ambas listas: números y símbolos. Es importante tener en cuenta que cuando generamos las permutaciones con "itertools.permutations", estamos creando un iterador, tipo generador, por lo que no todas las permutaciones son generadas cuando se le llama. La complejidad computacional de un algoritmo de permutación es $\mathcal{O}(n!)$.  Para medir los efectos reales (sin funciones de librerías) de la permutación, se ha usado una función **descargada directamente** de: https://stackoverflow.com/questions/52356091/recursive-algorithm-to-generate-all-permutations-of-length-k-of-a-list-in-python




````
import numpy as np
def permutacion(lst,n):
    # if length of permutation is negative or 0, return empty
    if n<=0:
        return [[]]
    # else take empty list l 
    l=[]
    # loop over whole lst
    for i in range(0,len(lst)): 
        m=lst[i] # current element of lst (ith) 
        remLst=lst[:i] + lst[i+1:]  # remaining list i.e. all elements except ith
        # recursive call to Perm with remaining list and decremented length 
        for p in permutacion(remLst,n-1): 
            # append current element + all sublists p generated through recursion as an item in list l
            l.append([m]+p) 
    return l

lst_numbers = [str(x) for x in range(1,10)]
permutacion(lst_numbers, 5);
````

````
import numpy as np
a = np.arange(1, 9)
sym = ['+', '-', '*', '/']
target = 6
while True:
  lst = np.random.choice(a, 5, replace=False)#seleccionamos 5 valores de la lista
  lst = [str(x) for x in lst]#convertimos a string los valores numéricos
  ls = [val for pair in zip(lst, sym) for val in pair] #combinamos ambas listas
  ls.append(lst[4])#puesto que no hay paridad de longitud, añadimos el último elemento
  formula = "".join(x for x in ls) #creamos la fórmula
  if eval(formula) == target: #evaluamos dicha fórmula 
    break #si se cumple, salimos del loop
print(formula)
````

Una alternativa sería aplicando el método de ramificación y poda. En este caso, consideramos siempre el mismo orden de operaciones: [* / + - ]. Tenemos pues:
$a*b/c+d-e = t$. La primera condición que ponemos es que las dos primeras operaciones resulten en un entero, y que el target sea mayor que el resutlado de la misma. Así, podemos sacar los otros valores tal que $d-e = t - (a*b/c)$ 

````
a = np.arange(1, 10)
target = 20 #Valor objetivo
while True: 
    n1 = np.random.choice(a, 3, replace=False) #Seleccionamos aleatoriamente 3 valores
    o1 = n1[0] * n1[1] / n1[2]  # calculamos la primera operación
    lst_aux = [x for x in a if x not in n1]# generamos una segunta lista que no contenga valores de la primera
    if n1[0] * n1[1] % n1[2] == 0 and abs(target-o1) != 0 and (np.max(lst_aux) - np.min(lst_aux)) >= abs(target-o1): #comprobamos que sea entero, que la diferencia no sea 0 y que en el resto de números podamos encontrar dicha diferencia
        #print(abs(target - o1))
        while True:
            n2 = np.random.choice(lst_aux, 2, replace=False) 
            #print(n2, '.........', n2[0] - n2[1])
            if (n2[0] - n2[1]) == abs(target - o1): #imponemos la condición para que se alcance el target
                break
        break
print(n1, n2)
````
