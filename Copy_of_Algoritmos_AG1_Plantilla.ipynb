```python
# N Reinas - Vuelta Atrás()
###############################################################################
# Verifica que en la solución parcial no hay amenazas entre reinas.
def solucion_prometedora(posicion_reinas, fila_nueva_reina):
  # Se actualiza la posición de las reinas considerando una nueva reina en la fila indicada.
  propuesta_posicion_reinas = posicion_reinas + [fila_nueva_reina]
  
  # Si hay dos valores iguales en la lista "posicion_reinas", significa que hay dos reinas en la misma fila.
  for columna in range(len(propuesta_posicion_reinas)):
    if propuesta_posicion_reinas.count(propuesta_posicion_reinas[columna]) > 1:
      return False
    
    # Se verifica que no haya dos reinas en la misma diagonal
    for otra_col in range(columna + 1, len(propuesta_posicion_reinas)):
      if abs(columna - otra_col) == abs(propuesta_posicion_reinas[columna] - propuesta_posicion_reinas[otra_col]):
        return False
  
  return True

# Dibuja la solución en un tablero.
def representar_solucion(num_solucion, posicion_reinas):
  n = len(posicion_reinas)
  
  print(f"Solución número {num_solucion}: {posicion_reinas}")
  print("-------------------------------\n")
  
  for fila in range(n):
    linea = ""
    
    for columna in range(n):
      if posicion_reinas[columna] == fila + 1:
        linea += " Q "
      else:
        linea += " - "
    
    print(linea)
  
  print("\n")

# Proceso principal de N-Reinas.
def resolver_N_reinas(N, posicion_reinas = [], num_soluciones = 0):
  # Se recorren todas las filas de la próxima columna a rellenar con una reina.
  for fila in range(N):
    # Si poner una reina en la fila actual rompe alguna restricción, se pasa a la siguiente fila.
    # En caso contrario, se sitúa una reina en la fila actual. Si ya tenemos N reinas, se muestra la solución, y si
    # no, pasamos a la siguiente columna llamando de forma recursiva a esta misma función.
    # Y en ambos casos, al acabar hay que eliminar la reina de la fila actual para seguir probando nuevas posiciones
    # en siguientes líneas.
    if solucion_prometedora(posicion_reinas,fila + 1):
      posicion_reinas += [fila + 1]
      
      if len(posicion_reinas) == N:
        num_soluciones += 1
        representar_solucion(num_soluciones, posicion_reinas)
      else:
        num_soluciones = resolver_N_reinas(N, posicion_reinas, num_soluciones)
      
      posicion_reinas.pop(-1)
    
  return num_soluciones
```

Algoritmo de fuerza bruta: los algoritmos de fuerza bruta enumeran todos los 
posibles candidatos y evalúan si estos cumplen con las condiciones de problema. 
Calculamos la distancia a pares entre cada uno de los puntos. 

El número de distancias posibles puede calcularse como el número de combinaciones sin repetición, de forma que: 

$\left( \begin{matrix} n \\ k\end{matrix} \right) = \frac{n!}{k!\,(n-k)!}$

En nuestro caso $n = 10^4, k = 2$

Puesto que se trata de un trabajo didáctico, se va a evitar el uso de funciones de librerías para calcular la diferencia:

```python
#Fuerza bruta
import numpy as np
def min_fuerza_bruta(lst):
  lst_diff = np.zeros((len(lst), len(lst))) #Inicialización de una lista con las diferencias por elemento
  for i in range(len(lst)):#Primer bucle que permite fijar el element sobre el que se calcula la diferencia
    for j in range(len(lst)):#Segundo bucle que permite leer la lista completa
      lst_diff[i][j] = abs(lst[i] - lst[j])#Cálculo de la diferencia

  min_val = np.min(lst_diff[np.nonzero(lst_diff)])#Búsqueda del minimo valor
  indx = np.where(lst_diff == min_val)#Índices de los valores que dan el valor mínimo
  return min_val

lst_1D = [1, 3, 4, 5, 9]
d = min_fuerza_bruta(lst_1D)
print('La mínima distancia es:', d)
```

----------------------------------------------------------------------------------------------------------------------------------------------------------------
Como se puede observar, la complejidad computacional de dicho experimento es de $\mathcal{O}(n^2)$.

Utilizamos ahora el paradigma de diseño "divide y vencerás".  Para el caso de 1D, podemos dividir el problema en las siguientes etapas:

1. Ordenar la lista de valores
2. Dividimos la lista en dos listas a patir de la mediana.
3. Calculamos la distancia mínima en cada una de las listas
4. Calculamos la distancia entre el último elemento de la primera lista y el primero de la segunda. 
5. Devolvemos el mínimo de las 3 distancias calculadas en los pasos (3) y (4). 






```python
def min_divide_venceras_1D(lst):
  
  if len(lst) <= 3:
    d = min_fuerza_bruta(lst)
    # print(lst, '.......', d)
    return d

  lst.sort()
  med = len(lst)//2
  lst_1 = lst[:med]
  lst_2 = lst[med:]
  # print('lista..........', lst)

  #Recursividad
  d1 = min_divide_venceras_1D(lst_1)
  d2 = min_divide_venceras_1D(lst_2)
  return min(d1, d2, abs(lst_1[-1] - lst_2[0]))


a = [1324, 464, 44, 55, 66, 89]
d = min_divide_venceras_1D(a);
print('La mínima distancia es:', d)

  
```
